```python
#Webscraping
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup as bs
import requests

import pickle
```

```python
driver=webdriver.Chrome(r"C:/chromedriver")
```

```python
#Opening page of OLX

driver.get("https://www.olx.in/cars_c84")
```

```python
#To get the urls of the Maharastra

Button=driver.find_element_by_xpath('//button[@class="rui-1rYgw"]')
```

```python
Button.send_keys('Maharashtra')
```

```python
#url

Maharashtra_car_urls=[]
for i in range(0,20):
    for j in driver.find_elements_by_xpath('//a[@class="fhlkh"]'):
          source=Maharashtra_car_urls.append(j.get_attribute('href'))
    try:        
        if(len(Maharashtra_car_urls)< 5000):
                print("start scrolling to generate more images on the page...")
                driver.execute_script("window.scrollBy(0,10000)")
                driver.find_element_by_xpath('//button[@class="rui-39-wj rui-3evoE rui-1JPTg"]').click()
                Maharashtra_car_urls.append(source)
                
        if(len(Maharashtra_car_urls)== 5000):
                        break
    except:
        print("no more records")              
        
          
```

```python
#Scrape the details for required attributes 
import time
Brand=[]
Model=[]
Variant=[]
Manufacturing_Year=[]
Driven_kilometres=[]
Fuel=[]
Number_of_owners=[]
Location=[]
Price=[]

for i in url:
    driver.get(i)
    time.sleep(2)
    try:
        Brand.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[1]/div/span[2]').text)
    except NoSuchElementException:
        Brand.append("-")
        
    try:
        Model.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[2]/div/span[2]').text)
    except NoSuchElementException:
        Model.append("-")
        
    try:
        Variant.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[3]/div/span[2]').text)
        
    except NoSuchElementException:
        Variant.append("-")
        
    try:
        Manufacturing_Year.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[4]/div/span[2]').text)
        
    except NoSuchElementException:
        Manufacturing_Year.append("-")
        
    try:
        Driven_kilometres.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[7]/div/span[2]').text)
        
    except NoSuchElementException:
        Driven_kilometres.append("-")
        
        
    try:
        Fuel.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[5]/div/span[2]').text)
        
    except NoSuchElementException:
        Fuel.append("-")
        
        
    try:
        Number_of_owners.append(driver.find_element_by_xpath('/html/body/div[1]/div/main/div/div/div/div[4]/section[1]/div/div/div[1]/div/div[8]/div/span[2]').text)
        
    except NoSuchElementException:
        Number_of_owners.append("-")
        
        
    try:
        Location.append(driver.find_element_by_xpath('//span[@class="_2FRXm"]').text)
        
    except NoSuchElementException:
        Location.append("-")
        
        
    try:
        Price.append(driver.find_element_by_xpath('//span[@class="_2xKfz"]').text)
        
    except NoSuchElementException:
        Price.append("-")
        
        
        
```
